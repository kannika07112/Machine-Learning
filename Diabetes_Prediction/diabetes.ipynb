```python
# Importing the numpy library, which helps with numerical operations (like working with arrays and matrices)
import numpy as np

# Importing pandas, a powerful library used for data manipulation and analysis (especially for tabular data)
import pandas as pd

# Importing StandardScaler from sklearn to standardize the features 
# (this means scaling data so it has a mean of 0 and a standard deviation of 1)
from sklearn.preprocessing import StandardScaler

# Importing a function to split the dataset into training and testing parts
# This helps us evaluate how well our model performs on unseen data
from sklearn.model_selection import train_test_split

# Importing the Support Vector Machine (SVM) classifier model
# This is the algorithm weâ€™ll use to train our model
from sklearn import svm

# Importing accuracy_score to check how well our model performed by comparing predictions to actual labels
from sklearn.metrics import accuracy_score
```

```python
# Reading the diabetes dataset from a CSV file using pandas
diabetes_dataset = pd.read_csv('diabetes.csv')

# Displaying the first 5 rows of the dataset to get an idea of what the data looks like
diabetes_dataset.head()
```

```python
#0-->non diabetic
#1-->diabetic
```

```python
# Getting the number of rows and columns in the dataset (rows = records, columns = features)
diabetes_dataset.shape
```

```python
# Getting statistical measures like mean, standard deviation, min, and max of each column
diabetes_dataset.describe()
```

```python
# Counting how many people are diabetic (1) and non-diabetic (0)
diabetes_dataset['Outcome'].value_counts()
#0-->non diabetic
#1-->diabetic
```

```python
# Getting the average values of all features grouped by the 'Outcome'
# This helps us compare feature differences between diabetic and non-diabetic people
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Separating the features and labels
# X contains all the input features (everything except the 'Outcome' column)
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
# Y contains the labels (the output we want to predict â€” 0 for non-diabetic, 1 for diabetic)
Y = diabetes_dataset['Outcome']
```

```python
# Displaying the feature set
print(X)
```

```python
# Displaying the labels
print(Y)
```

```python
# Data Standardization
# Initializing the StandardScaler to scale input features
scaler = StandardScaler()
# Fitting the scaler on the feature data (it calculates the mean and std dev for scaling)
scaler.fit(X)
```

```python
# Transforming the feature data into a standardized format (mean = 0, std = 1)
standardized_data = scaler.transform(X)
# Printing the standardized data
print(standardized_data)
```

```python
# Reassigning X with the scaled data
X = standardized_data

# Keeping the labels unchanged
Y = diabetes_dataset['Outcome']
```

```python
# Displaying the final standardized features and corresponding labels
print(X)
print('\n')
print(Y)
```

```python
# Splitting the data into training and testing sets
# 80% data for training, 20% for testing; random_state=2 ensures reproducibility
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
# Displaying the shapes of the full dataset, training set, and testing set
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Training the Support Vector Machine (SVM) Model
# Creating an SVM classifier with a linear kernel
classifier = svm.SVC(kernel='linear')
# Training the classifier using the training data
classifier.fit(X_train, Y_train)


# Model Evaluation

# Predicting labels for training data
X_train_prediction = classifier.predict(X_train)

# Calculating accuracy score for training data predictions
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score of the training data : ', training_data_accuracy)

# Predicting labels for test data
X_test_prediction = classifier.predict(X_test)

# Calculating accuracy score for test data predictions
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
# Making a Predictive System

# Input data: details of one person in the same order as the dataset features
input_data = (5,166,72,19,175,25.8,0.587,51)

# Converting the input data to a numpy array for model processing
input_data_as_numpy_array = np.asarray(input_data)

# Reshaping the input data to make it suitable for one prediction (1 sample, many features)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# Standardizing the input data using the same scaler used for training data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

# Making prediction using the trained SVM model
prediction = classifier.predict(std_data)
print(prediction)

# Interpreting the prediction result (0 = Not diabetic, 1 = Diabetic)
if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
